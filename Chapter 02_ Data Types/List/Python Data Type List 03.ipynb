Question 22: How to Remove Duplicates from a List While Preserving Order?

```python
a = [1, 2, 3, 4, 2, 3, 7, 6, 9, 5, 3]
unique_list = []
[unique_list.append(i) for i in a if i not in unique_list]
print(f"Unique List: {unique_list}")
```

```python
unique_list = []
for i in a:
    if i not in unique_list:
        unique_list.append(i)

print(f"Unique List: {unique_list}")
```

Question 24: How to Concatenate Multiple Lists into One?

Question 28: How to Sort a List Based on Another List?

```python
my_list1 = ["a", "b", "c"]
my_list2 = [2, 3, 1]
sortList = [i for _, i in sorted(zip(my_list2, my_list1))]
sortList
```
