Question 1: How to Create a List in Python?

```python
first_list = [2, 5.8, "Bytes of Intelligence"]
print(f"List is: {first_list}")
```

Question 2: How to Add an Element to a List?

```python
first_list = [2, 5.8, "Bytes of Intelligence"]
print(f"List is: {first_list}")
first_list.append("Dhaka")
print(f"After Append List is: {first_list}")
```

Question 3: How to Remove an Element from a List?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka']
print(f"List is: {first_list}")
first_list.remove(5.8)
print(f"After Append List is: {first_list}")
```

Question 4: How to Access Elements in a List?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka']
print(f"List Element is: {first_list[1]}")
```

Question 5: How to Concatenate Two Lists?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka']
li2 = ["Cumilla", 1, 2, 3, 4]
first_list_add = first_list + li2
print(f"List is: {first_list_add}")
```

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka']
li2 = ["Cumilla", 1, 2, 3, 4]
first_list_add = li2 + first_list
print(f"List is: {first_list_add}")
```

Question 6: How to Find the Length of a List?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka', 3, 4, 5, 6]
print(f"Len of List is: {len(first_list)}")
```

Question 7: How to Sort a List?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka', 3, 4, 5, 6]
first_list.sort()
print(f"Len of List is: {first_list}")
```

```python
first_list = [10, 6, 2, 4, 0, 12, 2, 5.8, 3, 4, 5, 6]
first_list.sort()
print(f"Len of List is: {first_list}")
```

Question 8: How to Reverse the Order of Elements in a List?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka', 3, 4, 5, 6]
print(f"List is: {first_list[::-1]}")
#madam
```

Question 9: How to Slice a List in Python?

```python
first_list = [2, 5.8, 'Bytes of Intelligence', 'Dhaka', 3, 4, 5, 6]
print(f"List is: {first_list[0:6]}")
#madam
```
